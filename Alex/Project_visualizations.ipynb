```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
comparison_data = "Resources/comparison.csv"
```

```python
comparison_df = pd.read_csv(comparison_data)
comparison_df.head()
```

```python
#Top ten state overdose rates in 2019
overdose2019_top10 = comparison_df.sort_values(by=['Overdose Rate 2019'], ascending=False).head(10)

#Top ten state overdose rates in 2018
overdose2018_top10 = comparison_df.sort_values(by=['Overdose Rate 2018'], ascending=False).head(10)

#Top ten state overdose rates in 2017
overdose2017_top10 = comparison_df.sort_values(by=['Overdose Rate 2017'], ascending=False).head(10)

#Top ten state overdose rates in 2016
overdose2016_top10 = comparison_df.sort_values(by=['Overdose Rate 2016'], ascending=False).head(10)

#Top ten state overdose rates in 2015
overdose2015_top10 = comparison_df.sort_values(by=['Overdose Rate 2015'], ascending=False).head(10)
```

```python
overdose2019_top10
```

```python
# Cleaner dfs
#Top ten state overdose rates in 2019
top102019 = overdose2019_top10[["State", "Overdose Rate 2019"]]
top102019.reset_index(drop=True)

#Top ten state overdose rates in 2018
top102018 = overdose2018_top10[["State", "Overdose Rate 2018"]]
top102018.reset_index(drop=True)

#Top ten state overdose rates in 2017
top102017 = overdose2017_top10[["State", "Overdose Rate 2017"]]
top102017.reset_index(drop=True)

#Top ten state overdose rates in 2016
top102016 = overdose2016_top10[["State", "Overdose Rate 2016"]]
top102016.reset_index(drop=True)

#Top ten state overdose rates in 2015
top102015 = overdose2015_top10[["State", "Overdose Rate 2015"]]
top102015.reset_index(drop=True)
```

```python
state = top102019["State"]
rates = top102019["Overdose Rate 2019"]

x_axis = np.arange(len(rates))

plt.bar(x_axis, rates, color="b", alpha = .5, align="center")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, state, rotation='vertical')

plt.title("Top Ten Overdose Rate States in 2019")
plt.ylabel("Overdose Rate")
plt.xlabel("State")
plt.show()

```

```python
state = top102018["State"]
rates = top102018["Overdose Rate 2018"]

x_axis = np.arange(len(rates))

plt.bar(x_axis, rates, color="Red", alpha = .5, align="center")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, state, rotation='vertical')

plt.title("Top Ten Overdose Rate States in 2018")
plt.ylabel("Overdose Rate")
plt.xlabel("State")
plt.show()
```

```python
overdose_rates = top102019["Overdose Rate 2019"]
fig1, ax1 = plt.subplots()
ax1.set_title('Overdose Rates by State in the US')
ax1.set_ylabel('Percent of the Population')
ax1.boxplot(overdose_rates)
plt.show()

overdose_rates = top102018["Overdose Rate 2018"]
fig1, ax1 = plt.subplots()
ax1.set_title('Overdose Rates by State in the US')
ax1.set_ylabel('Percent of the Population')
ax1.boxplot(overdose_rates)
plt.show()
```
